```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print("Dataset Head (first 10 rows):")
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

```python
# 1. Total sales by region
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()
print("Total sales by region:\n", region_sales)

# 2. Transaction count by category
category_counts = df.groupby('category')['transaction_id'].count().reset_index(name='transaction_count')
print("\nTransactions per category:\n", category_counts)

# 3. Average sales per salesperson
salesperson_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()
print("\nAvg sales per salesperson:\n", salesperson_avg)

# 4. Sort regional sales descending
sorted_region_sales = region_sales.sort_values(by='sales_amount', ascending=False)
print("\nSorted regional sales (top performer highlighted):\n", sorted_region_sales)
print("\nTop-performing region:", sorted_region_sales.iloc[0]['region'])
```

```python
# 1. Group by region + category
region_category_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()
print("Sales by region & category:\n", region_category_sales)

# 2. Multiple aggregations per salesperson
salesperson_metrics = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()
print("\nSalesperson metrics:\n", salesperson_metrics)

# 3. Sort by total sales
sorted_salesperson = salesperson_metrics.sort_values(by='sum', ascending=False)
print("\nTop salesperson by total sales:\n", sorted_salesperson)
print("\nTop performer:", sorted_salesperson.iloc[0]['salesperson'])

# 4. Category with max revenue
category_total = df.groupby('category')['sales_amount'].sum()
print("\nHighest-revenue category:", category_total.idxmax(), f"(${category_total.max():.2f})")
```

```python
# Custom function
def sales_range(x):
    return x.max() - x.min()

# Advanced aggregation by region
region_advanced = df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
print("Region sales with custom range:\n", region_advanced)

# Final summary report (multi-level columns)
summary_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()
print("\nFinal regional summary report:\n", summary_report)

# Explanation of column structure
print("\nMulti-level column explanation:")
print("• Dictionary agg() creates a MultiIndex (tuples): e.g. ('sales_amount', 'sum')")
print("• Single aggregations produce flat columns.")
print("• To flatten: summary_report.columns = ['region', 'total_sales', 'avg_sales', 'transaction_count']")
```
