```python
import numpy as np

# Given data
data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])

print("Task 1")
print("───────")

# Print shape of the array
print("Shape of data:", data.shape)

# Mean temperature per station (mean along axis=1 → per row)
mean_per_station = data.mean(axis=1)
print("\nMean temperature per station:")
for i, mean_temp in enumerate(mean_per_station, 1):
    print(f"Station {i}: {mean_temp:.2f} °C")

print("\nTask 2")
print("───────")

# Boolean mask: temperatures > 28.0
mask = data > 28.0

# Extract matching values and flatten to 1D
high_temps = data[mask]
print("Temperatures above 28.0°C (as 1D array):")
print(high_temps)

print("\nTask 3")
print("───────")

# Normalize to [0, 1]
data_min = data.min()
data_max = data.max()
normalized = (data - data_min) / (data_max - data_min)

# Round to 2 decimal places and print
print("Normalized data (rounded to 2 decimal places):")
print(np.round(normalized, 2))
```
